```python
import os, sys
from PIL import Image
import numpy as np
from skimage.measure import shannon_entropy
import imageio
```

```python
pathA = "data/alzheimers"
filenamesA = os.listdir(pathA)
EA = []

for name in filenamesA:
    img = imageio.imread(pathA+"/"+name)
    img = np.asarray(img)
    e = shannon_entropy(img)
    EA.append(e)
```

```python
pathB = "data/nonalzheimers"
filenamesN = os.listdir(pathB)
EN = []

for name in filenamesN:
    img = imageio.imread(pathB+"/"+name)
    img = np.asarray(img)
    e = shannon_entropy(img)
    EN.append(e)
```

```python
EA = np.array(EA)
EN= np.array(EN)
```

```python
dA = {}
for file, e in zip(filenamesA, EA):
    dA[file] = e
```

```python
dN = {}
for file, e in zip(filenamesN, EN):
    dN[file] = e
```

```python
sorted_values = sorted(dA.values()) # Sort the values
sortedA = {}

for i in sorted_values:
    for k in dA.keys():
        if dA[k] == i:
            sortedA[k] = dA[k]
            break

print(sortedA)
```

```python
sorted_values = sorted(dN.values()) # Sort the values
sortedN = {}

for i in sorted_values:
    for k in dN.keys():
        if dN[k] == i:
            sortedN[k] = dN[k]
            break

print(sortedN)
```
