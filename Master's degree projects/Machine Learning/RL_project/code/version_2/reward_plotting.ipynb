# Analyze the progress of the training and printing the trends 

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
%matplotlib inline
plt.rcParams["figure.figsize"] = (12,6)

plt.style.use('ggplot')
```

```python
# Update the values as in the trainings.
date = "Feb-04_14:30:07"
directory = "immagini/"
epis = "epis1000_"
gamma = "gam0.99_"
extraInfo = "_leakyReLU_Entr0.4_decay-0.008"
```

```python
directory = directory+date+"---"+epis+gamma+extraInfo+"/"
if not(os.path.exists(directory)):
    os.makedirs(directory)
```

```python
def df_plot(dfs, x, ys, ylim=None, legend_loc='best'):
    if ylim is not None:
        plt.ylim(ylim)
    for df, name in dfs:
        name = name.split('_')[1]
        for y in ys:
            plt.plot(df[x], df[y], linewidth=2,
                     label=name + ' ' + y.replace('_', ''))
    plt.xlabel(x.replace('_', ''))
    plt.legend(loc=legend_loc)
    plt.savefig(directory+ys[0]+".png")
    plt.show()
```

```python
# ENTER LIST OF LOG FILENAMES HERE:
filepaths = 'log-files/CartPole-v1/'+date+'/log.csv'
dataframes = [pd.read_csv(filepaths)]
names = [filepaths]
data = list(zip(dataframes, names))
```

# Plots

```python
fig = plt.figure()
fig = df_plot(data, '_Episode', ['_MeanReward'])
```
